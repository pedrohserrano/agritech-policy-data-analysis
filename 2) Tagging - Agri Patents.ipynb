```python
from flashtext import KeywordProcessor
import time
import pandas as pd
import warnings
warnings.filterwarnings("ignore")

terms = ['agriculture','agricultural','agritech','agtech','agronomy','farming','gardening','agronomics','agrology','horticulture','agribusiness','cultivation', 'forestry', 'animal husbandry', 'hunting', 'fishing', 'planting', 'sowing', 'fertisiling', 'harvesting', 'mowing','aviculture', 'apiculture', 'pisciculture', 'sowing', 'planting', 'manuring', 'harvesting']
print('terms to use: {}'.format(len(terms)))
data_patents = pd.read_excel('datasets/patents_ccm_28072020.xlsx')
data_patents = data_patents[data_patents['count_matches'] > 0]
data_patents.head(4)
```

---
## Tagging CCM Patents related to Agriculture

```python
def extract(vec, dictionary, info=False):
    matrix = []
    for line in vec:
        matrix.append(dictionary.extract_keywords(str(line), span_info=info))
    return matrix
```

```python
terms_dict = KeywordProcessor()
terms_dict.add_keywords_from_list(terms)
terms_extracted = extract(data_patents.abstract, terms_dict)
rows = [list(set(i)) if len(i)>0 else '' for i in terms_extracted]
data_patents['matches'] = [str(i).replace('[', '').replace(']', '') for i in rows]
data_patents['count_matches'] = [len(i) for i in terms_extracted]
```

---
## Dummy for year as from 1992

```python
#  extracting the year from the date column
data_patents['First_year']=data_patents['date'].str.slice(stop=4)

# change year to numeric column
data_patents["First_year"] = pd.to_numeric( data_patents["First_year"],downcast='integer')
```

```python
# creating dummy for all laws passed starting with the foundation of the EU in 1992
After_1992= []
for i in data_patents.First_year:
    year =  1 if i >1991 else 0
    After_1992.append(year)

# adding the dummy to the original dataset
data_patents['After_1992']=After_1992
data_patents.After_1992.sum()
```
